```python
# ===============================================================
#   CALCULADORA B√ÅSICA E ESTRUTURADA ‚Äî PYTHON / GOOGLE COLAB
#   Autora: Mariana Diniz de Oliveira
#   Fun√ß√µes: soma, subtra√ß√£o, multiplica√ß√£o, divis√£o, pot√™ncia,
#            m√≥dulo, m√©dia, raiz quadrada
#   Requisitos da EBAC atendidos + c√≥digo limpo e organizado
# ===============================================================
import math

# -----------------------------------------------------------
# Fun√ß√£o para ler o modo num√©rico
# -----------------------------------------------------------
def ler_modo_numero():
    """
    L√™ 'int' ou 'float'. Se o usu√°rio apertar ENTER, usa float.
    Aceita somente 'int', 'float' ou vazio.
    """
    while True:
        modo = input("Modo num√©rico ('int' ou 'float') [ENTER = float]: ").strip().lower()

        if modo == "":
            print("‚Üí Modo definido como FLOAT")
            return "float"

        if modo in ("int", "float"):
            print(f"‚Üí Modo definido como {modo.upper()}")
            return modo

        print("Entrada inv√°lida. Digite apenas 'int', 'float' ou aperte ENTER.")


# -----------------------------------------------------------
# Convers√£o segura
# -----------------------------------------------------------
def converter_valor(texto, modo):
    while True:
        entrada = input(texto).strip().replace(",", ".")
        try:
            return int(float(entrada)) if modo == "int" else float(entrada)
        except ValueError:
            print("Valor inv√°lido. Digite um n√∫mero.")


# -----------------------------------------------------------
# L√™ dois n√∫meros
# -----------------------------------------------------------
def ler_dois_numeros(modo):
    a = converter_valor("Primeiro n√∫mero: ", modo)
    b = converter_valor("Segundo n√∫mero: ", modo)
    return a, b


# -----------------------------------------------------------
# L√™ v√°rios n√∫meros para m√©dia
# -----------------------------------------------------------
def ler_lista_numeros(modo):
    lista = []
    print("Digite valores para m√©dia (x para parar):")

    while True:
        entrada = input("Valor: ").strip().replace(",", ".")

        if entrada.lower() == "x":
            break

        try:
            valor = int(float(entrada)) if modo == "int" else float(entrada)
            lista.append(valor)
        except ValueError:
            print("Valor inv√°lido.")

    return lista


# -----------------------------------------------------------
# Opera√ß√µes matem√°ticas
# -----------------------------------------------------------
def soma(a, b): return a + b
def subtracao(a, b): return a - b
def multiplicacao(a, b): return a * b
def divisao(a, b): return "Erro: divis√£o por zero" if b == 0 else a / b
def potencia(a, b): return a ** b
def modulo(a, b): return "Erro: divis√£o por zero" if b == 0 else a % b
def media(lista): return "Erro: lista vazia" if len(lista) == 0 else sum(lista) / len(lista)
def raiz(a): return "Erro: raiz de negativo" if a < 0 else math.sqrt(a)


# -----------------------------------------------------------
# Menu
# -----------------------------------------------------------
def mostrar_menu():
    print("\n" + "="*50)
    print("CALCULADORA ‚Äî MENU")
    print("="*50)
    print("1) Soma")
    print("2) Subtra√ß√£o")
    print("3) Multiplica√ß√£o")
    print("4) Divis√£o")
    print("5) Pot√™ncia")
    print("6) Resto (m√≥dulo)")
    print("7) M√©dia de valores")
    print("8) Raiz quadrada")
    print("9) Alterar modo num√©rico")
    print("0) Sair")
    print("="*50)


# -----------------------------------------------------------
# Programa Principal
# -----------------------------------------------------------
def main():
    print("="*60)
    print("üî¢ CALCULADORA ‚Äî PYTHON / GOOGLE COLAB")
    print("="*60)

    modo = ler_modo_numero()
    rodando = True

    while rodando:
        mostrar_menu()
        opc = input("Escolha: ").strip()

        if opc == "0":
            print("Saindo...")
            break

        elif opc == "1":
            a, b = ler_dois_numeros(modo)
            print("Resultado:", soma(a, b))

        elif opc == "2":
            a, b = ler_dois_numeros(modo)
            print("Resultado:", subtracao(a, b))

        elif opc == "3":
            a, b = ler_dois_numeros(modo)
            print("Resultado:", multiplicacao(a, b))

        elif opc == "4":
            a, b = ler_dois_numeros(modo)
            print("Resultado:", divisao(a, b))

        elif opc == "5":
            a, b = ler_dois_numeros(modo)
            print("Resultado:", potencia(a, b))

        elif opc == "6":
            a, b = ler_dois_numeros(modo)
            print("Resultado:", modulo(a, b))

        elif opc == "7":
            lista = ler_lista_numeros(modo)
            print("M√©dia:", media(lista))

        elif opc == "8":
            a = converter_valor("N√∫mero: ", modo)
            print("Raiz:", raiz(a))

        elif opc == "9":
            modo = ler_modo_numero()

        else:
            print("Op√ß√£o inv√°lida.")

        print("-"*50)

    print("Programa finalizado.")


# Executar
main()
```
